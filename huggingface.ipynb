#### Sentence Transformers on hugging face

###### Hugging face sentence-transformers is a python framework for state-of-the-art sentence, text and image embeddings. One of the embedding models is used in HuggingFaceEmbeddings class. We have also added alias for SentenceTransformerEmbeddings for users who are more familiar with directly using that package

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
text = "this is atest documents"
query_result = embeddings.embed_query(text)
len(query_result)
```

```python
doc_result = embeddings.embed_documents([text, "This is not a test document"])
doc_result[0]
```
